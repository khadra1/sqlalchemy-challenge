```python
import pandas as pd
from datetime import datetime as dt
from scipy import stats as st
import numpy as np
from numpy import mean
```

```python
# "tobs" is "temperature observations"
stations_tobs = pd.read_csv('../Resources/hawaii_measurements.csv')
stations_tobs.head()
```

```python
# Checking date column type before converting to datetime
stations_tobs.info()
```

```python
# Convert the date column format from string to datetime
stations_tobs['date']= pd.to_datetime(stations_tobs['date'])

# Checking date column type after converting to datetime
stations_tobs.info()
stations_tobs.head()
```

```python
# Set the date column as the DataFrame index
stations_tobs = stations_tobs.set_index('date')
stations_tobs.head()
```

```python
# Drop the date column
stations_obs = stations_tobs.reset_index(drop=False)
stations_obs.head()
```

### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Filter data for desired months
# Filter data for desired months
june_tobs = stations_obs[stations_obs['date'].dt.month == 6]
june_tobs = june_tobs.reset_index(drop=True)
dec_tobs = stations_obs[stations_obs['date'].dt.month == 12]
dec_tobs = dec_tobs.reset_index(drop=True)
```

```python
# Identify the average temperature for June
june_avg = june_tobs.groupby(['station']).mean()
june_avg
```

```python
# Identify the average temperature for December
dec_avg = dec_tobs.groupby(['station']).mean()
dec_avg
```

```python
# Run paired t-test on December and June Temperatures
st.ttest_rel(june_avg['tobs'], dec_avg['tobs'])
```

## June and December Temperature Analysis
#### I ran a paired t-test to find out if there is a meaningful relationship between the 2 months' temperatures, a paired t-test because I'm comparing the means of 2 months' temperatures in the same group (same stations):

#### With a p-value of .0001 the data is statistically significant (showing a meaningful relationship).
